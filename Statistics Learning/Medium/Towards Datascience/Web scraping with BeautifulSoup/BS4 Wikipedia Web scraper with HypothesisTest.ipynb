```python
Epidemiology_url='https://en.wikipedia.org/wiki/Epidemiology_of_depression'
```

```python
Epidemiology_wiki_html=urlopen(Epidemiology_url)
```

```python
Epidemiology_soup=BeautifulSoup(Epidemiology_wiki_html,'html.parser')
```

##  Data

### Data Scraping 

### Looking for the 'table' tag in the entire document

```python
Epidemiology_table=Epidemiology_soup.find_all('table')
```

### Data Wrangling

```python
import re
def num_processing(num):
    return float(re.sub(r'[^\w\s.]','',num))
```

```python
num_processing('156.30')
num1 = float(re.sub(r'[^\w\s.]','','1,156.30'))
num1
```

```python
# Creating an array for holding the extracted data
ranks=[]
rates=[]
countries=[]
for table in Epidemiology_table:
    rows=table.find_all('tr')
    for row in rows:
        cells=row.find_all('td')
        
        if len(cells)>1:
            rank=cells[0]
            ranks.append(rank.text)
            
            country=cells[1]
            countries.append(country.text.strip())
            
            rate=cells[2]
            rates.append(num_processing(rate.text.strip()))
    
```

```python
epidemiology_df = pd.DataFrame(ranks, index= countries)
#columns = ['Rank'])df1['DALY rate'] = rates
```
